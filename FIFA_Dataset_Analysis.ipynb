```python
#plotting a bar graph for no.of players wrt to their nationalities
x=fifa['nationality'].value_counts()[0:5].keys()
y=fifa['nationality'].value_counts()[0:5]
plt.figure(figsize=(5,5))
plt.bar(x,y,color="r")
plt.show()
```

```python
player_salary=fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

```python
#plotting a bar graph of players wrt to their salries

p1=list(player_salary['short_name'])[0:5]
p2=list(player_salary['wage_eur'])[0:5]
plt.figure(figsize=(7,7))
plt.bar(p1,p2,color=["blue","green","yellow","red","pink"])
plt.show()
```
